```python
# 선형회귀
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd

X1 = df['type_몰입']
y = df['사용의도']
X2 = df[['fun_감각'	,'fun_조작'	,'fun_도전']]
```

```python
X1, X2 , y
```

```python
X1.shape , X2.shape , y.shape
```

```python
lr = LinearRegression()
```

```python
# 데이터 분할
from sklearn.model_selection import train_test_split
```

```python
X1_train, X1_test, X2_train, X2_test, y_train, y_test = train_test_split(X1, X2, y, test_size=0.25, random_state=42)
```

```python
X1_train, X1_test, X2_train, X2_test, y_train, y_test
```

```python
X1_train.shape, X1_test.shape, X2_train.shape, X2_test.shape, y_train.shape, y_test.shape
```

```python
model_1 = lr.fit(X1_train.values.reshape(-1,1), y_train)
```

```python
X1_train.values.reshape(-1,1)
```

```python
model_1.score(X1_test.values.reshape(-1,1), y_test)
```

```python
import statsmodels.api as sm  ## OLS 분석을 보고 싶으면 statsmodels 을 따로 사용해야 된다.
## sklearn 에서 수행한 결과와 동일

x = sm.add_constant(X1_train)
model = sm.OLS(y_train,x)
result = model.fit()
print(result.summary())
```

```python
import matplotlib.pyplot as plt

plt.scatter(X1_train, y_train)
plt.scatter(X1_test, y_test)
plt.plot(X1_train, model_1.predict(X1_train.values.reshape(-1,1)), color='red')
plt.plot(X1_test, model_1.predict(X1_test.values.reshape(-1,1)), color='blue')
plt.legend(['train', 'test'])
plt.show()
```
